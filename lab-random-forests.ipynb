```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load datasets
categorical = pd.read_csv('./files_for_lab/categorical.csv')
numerical = pd.read_csv('./files_for_lab/numerical.csv')
target = pd.read_csv('./files_for_lab/target.csv')
```

```python
# Show categorical data
categorical.head()
```

```python
# Encode categorical data
from sklearn.preprocessing import OneHotEncoder

# Instantiate encoder
encoder = OneHotEncoder()

# Fit encoder
encoder.fit(categorical)

# Encode X_train_cat_nom and X_test_cat_nom with fitted encoder. 
X_cat_encoded = encoder.transform(categorical).toarray()

# Get feature names from encoder
column_names = encoder.get_feature_names_out()

# Convert np.array to DataFrame by adding columns
X_cat_encoded_df = pd.DataFrame(X_cat_encoded, 
                                columns=column_names, 
                                index=categorical.index) # <-IMPORTANT- keeps index order for concat
```

```python
# Concatenate dfs
X = pd.concat([numerical, X_cat_encoded_df], axis=1)

# Set target value y
y = target['TARGET_B']
```

```python
# Store target value counts
count_classes = y.value_counts()

# Print class imbalance ratio
print("The class imbalance ratio is: {:.2f}".format((count_classes[0]-count_classes[1])/(count_classes[0]+count_classes[1])))
print(count_classes)
```

+ Theres is class imbalance, apply **RandomOverSampler** to balance clases

```python
from sklearn.model_selection import train_test_split

# Train-test split data
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0) # default test_size value is .2
```

```python
from imblearn.over_sampling import RandomOverSampler 

# Load RandomOverSampler
ros = RandomOverSampler(random_state=42)

# Resample train data
X_train_over, y_train_over = ros.fit_resample(X_train, y_train)

# Print balanced value counts
y_train_over.value_counts()
```

```python
# Scale data
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X_train_over)

X_train_over_scaled_np = scaler.transform(X_train_over)
X_test_scaled_np  = scaler.transform(X_test)

X_train_over_scaled_df = pd.DataFrame(X_train_over_scaled_np)
X_test_scaled_df  = pd.DataFrame(X_test_scaled_np)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix

clf = RandomForestClassifier(max_depth=5,
                             min_samples_split=20,
                             min_samples_leaf =20,
                             max_samples=0.8,
                             random_state = 42)

clf.fit(X_train_over_scaled_df, y_train_over)

print("The Accuracy for the Random Forest in the TRAIN set is {:.2f}".format(clf.score(X_train_over_scaled_df, y_train_over)))
print("The Accuracy for the Random Forest in the TEST  set is {:.2f}".format(clf.score(X_test_scaled_df, y_test)))

y_pred = clf.predict(X_test_scaled_df)

display(y_test.value_counts())
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm_test = confusion_matrix(y_test,y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm_test, display_labels=clf.classes_)
disp.plot()
plt.show()
```

+ In this business case, the target variable y tries to predict if a donor will donate or not.
Let's analize the different scenarios:
1. False positive -> model predicts a donor will donate but he doesn't
2. False negative -> model predicts a donor will not donate but he does

+ If we are talking about donating money, the False negative will have a much higher cost for the business

![image](https://miro.medium.com/v2/resize:fit:720/format:webp/1*UO_rS1oFtykUmriAWphUqQ.png)

When false negatives are of higher concern than false positives, we want to optimize for Recall:
+ recall = True positives / True positives + False negatives


```python
from sklearn.metrics import make_scorer, recall_score
from sklearn.model_selection import GridSearchCV

recall_scorer = make_scorer(recall_score)

param_grid = {
    'max_depth': [5, 7], # 3 * 2 * 2 * 2 * 1 = 3 * 8 = 24
    'min_samples_split': [5, 10],
    'min_samples_leaf' : [5, 10],
    'max_samples':[0.6, 0.8]
    }

rf_classifier = RandomForestClassifier(random_state=100)

grid_search = GridSearchCV(rf_classifier, param_grid, cv=5, return_train_score=True, n_jobs=-1, verbose = 20, scoring=recall_scorer)
grid_search.fit(X_train_over_scaled_df, y_train_over)
grid_search.best_params_ #To check the best set of parameters returned
```
